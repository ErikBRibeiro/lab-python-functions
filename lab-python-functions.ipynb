In the previous exercise, you improved the code for managing customer orders by using loops and flow control. Now, let's take it a step further and refactor the code by introducing functions.

Follow the steps below to complete the exercise:

2. Define a function named `get_customer_orders` that takes no parameters. Inside the function, implement the code for prompting the user to enter the product names using a loop. The function should return the `customer_orders` set.

3. Define a function named `update_inventory` that takes `customer_orders` and `inventory` as parameters. Inside the function, implement the code for updating the inventory dictionary based on the customer orders.

4. Define a function named `calculate_order_statistics` that takes `customer_orders` and `products` as parameters. Inside the function, implement the code for calculating the order statistics (total products ordered, and percentage of unique products ordered). The function should return these values.

5. Define a function named `print_order_statistics` that takes `order_statistics` as a parameter. Inside the function, implement the code for printing the order statistics.

```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
number_of_products = []

def initialize_inventory(products):
    inventory = {}
    for i in range(0, 5):
        inventory.update ({products[i] : int(input(f'How many {products[i]} do you have in inventory?'))})
    return (inventory)

def get_customer_orders(products):
    customer_orders = set()
    add_more = 'YES'
    while True:
        order = input('Insert the name of a product within the list that the customer wants to order: ').lower()
            
        if order in products:
            customer_orders.add(order)
            while add_more != 'NO':            
                add_more = input('Do you want to add another order? (YES/NO): ').upper()
                if add_more == 'YES':
                    break
                if add_more == 'NO':
                    print('Processing orders...')
                    return customer_orders
                elif add_more != 'YES':
                    print('Invalid input, try again.')
        else:
            print('The product is not in the list. Please try again.')
            
def update_inventory(customer_orders, inventory):
    # First, update the inventory based on customer orders
    for item in customer_orders:
        if item in inventory:
            inventory[item] -= 1

    # Now, filter out items with a quantity of zero using comprehension
    inventory = {key: value for key, value in inventory.items() if value > 0}

    return inventory

def calculate_order_statistics(customer_orders, products, inventory):
    total_products_ordered = len(customer_orders)
    total_inventory_list = list(inventory.values())
    total_inventory = 0
    for i in total_inventory_list:
        total_inventory +=i
    percentage_ordered = total_products_ordered / total_inventory * 100
    unique_percentage_ordered = len(customer_orders) / len(products) * 100
    order_statistics = []
    order_statistics.append(total_products_ordered)
    order_statistics.append(percentage_ordered)
    order_statistics.append(unique_percentage_ordered)
    print_order_statistics (order_statistics)

def print_order_statistics (order_statistics):
    print('Order Statistics:')
    print(f'Total Products Ordered: {order_statistics[0]}')
    print(f'The order was {round(order_statistics[1], 1)}% of all our products')
    print(f'The order includes at least 1 of {order_statistics[2]}% of our items')

def print_updated_inventory(inventory, products):
    for product in products:
        if product in inventory:
            print(f'Current {product} on inventory: {inventory[product]}')
        else:
            print(f'{product} is out of stock')

def get_prices_for_ordered_items(customer_orders):
    prices = {}
    for item in customer_orders:
        # Asking for the price of each item and converting it to a float
        price = float(input(f'Enter the price of {item}: '))
        prices[item] = price
    return prices

def print_total_price(customer_orders, prices):
    total_price = sum(prices[item] for item in customer_orders if item in prices)
    print(f'Total Price of the Order: ${total_price:.2f}')

inventory = initialize_inventory(products)
customer_orders = get_customer_orders(products)
inventory = update_inventory(customer_orders,inventory)
order_statistics = calculate_order_statistics(customer_orders, products, inventory)
print_updated_inventory (inventory,products)
prices = get_prices_for_ordered_items(customer_orders)
print_total_price(customer_orders, prices)
```
